```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains.summarize import load_summarize_chain
from langchain import PromptTemplate
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
load_dotenv()
api_key = os.getenv("GROQ_API_KEY")

loader = PyPDFLoader('apjspeech.pdf')
docs = loader.load_and_split()
docs
```

```python
final_documents = RecursiveCharacterTextSplitter(chunk_size =2000,chunk_overlap=100).split_documents(docs)
final_documents
```

```python
chunks_prompt = """
Please Summarize the below speech:
Speech: {text}
Summary:
"""

map_prompt_template = PromptTemplate(input_variables=['text'],
                                     template=chunks_prompt)
```

```python
final_prompt = """
Provide the final summary of entire speech with these important points.
Add a Motiation Title, Start the Precise summary with an introduction and provide the summary in number points for the speech.
Speech:{text}
"""

final_prompt_template = PromptTemplate(input_variables = ['text'],
                                       template=final_prompt)

final_prompt_template
```

```python
llm = ChatGroq(groq_api_key = api_key,model="Gemma-7b-It")
```

```python
summary_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt = map_prompt_template,
    combine_prompt = final_prompt_template,
    verbose=True
)

output = summary_chain.run(final_documents)
output
```
